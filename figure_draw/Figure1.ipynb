```python
# -*- coding: utf-8 -*-
"""
Created on Mon Nov 30 09:35:23 2020

@author: Administrator
"""
import matplotlib.pyplot as plt
plt.rcParams['savefig.dpi'] = 300 # 图片像素
plt.rcParams['figure.dpi'] = 300 # 分辨率
list1=list(range(1,10))
list2=[1,
       2,
       7,
       7.5,
       5,
       5,
       4.5,
       5.5,
       5]
plt.figure(figsize=(12,6))
plt.text(1,1,'1',ha='right',va= 'bottom',fontsize=19)
plt.text(9,5,'2',ha='right',va= 'bottom',fontsize=19)
plt.scatter([1,9],[1,5],color='red',s=50)

p1 = plt.plot(list1,list2,color='#836FFF',linewidth=2,label='raw data')

p2 = plt.plot([1,9],[1,5],'--',color='#76EE00',linewidth=2)
plt.grid(axis="y")
plt.legend([p1,p2],labels=['raw_data_line','segment_line'],prop={'size':20},loc='lower right')
plt.xticks(list(range(10)),['','1', '2', '3', '4', '5', '6', '7', '8', '9'],fontsize=20)
plt.yticks(list(range(8)),['','1', '2', '3', '4', '5', '6', '7'],fontsize=20)
plt.show()
#第一个中位数点
plt.figure(figsize=(12,6))
plt.text(1,1,'1',ha='right',va= 'bottom',fontsize=19)
plt.text(5,5,'3',ha='right',va= 'bottom',fontsize=19)
plt.text(9,5,'2',ha='right',va= 'bottom',fontsize=19)
plt.scatter([1,5,9],[1,5,5],color='red',s=50)

p1 = plt.plot(list1,list2,color='#836FFF',linewidth=2,label='raw data')

p2 = plt.plot([1,5,9],[1,5,5],'--',color='#76EE00',linewidth=2)
plt.grid(axis="y")
plt.legend([p1,p2],labels=['raw_data_line','segment_line'],prop={'size':20},loc='lower right')
plt.xticks(list(range(10)),['','1', '2', '3', '4', '5', '6', '7', '8', '9'],fontsize=20)
plt.yticks(list(range(8)),['','1', '2', '3', '4', '5', '6', '7'],fontsize=20)
plt.show()
# =============================================================================
plt.figure(figsize=(12,6))
plt.text(1,1,'1',ha='right',va= 'bottom',fontsize=19)
plt.text(5,5,'3',ha='right',va= 'bottom',fontsize=19)
plt.text(9,5,'2',ha='right',va= 'bottom',fontsize=19)
plt.text(2,2,'4',ha='right',va= 'bottom',fontsize=19)
plt.scatter([1,2,5,9],[1,2,5,5],color='red',s=50)

p1 = plt.plot(list1,list2,color='#836FFF',linewidth=2,label='raw data')

p2 = plt.plot([1,5,9],[1,5,5],'--',color='#76EE00',linewidth=2)
plt.grid(axis="y")
plt.legend([p1,p2],labels=['raw_data_line','segment_line'],prop={'size':20},loc='lower right')
plt.xticks(list(range(10)),['','1', '2', '3', '4', '5', '6', '7', '8', '9'],fontsize=20)
plt.yticks(list(range(8)),['','1', '2', '3', '4', '5', '6', '7'],fontsize=20)
plt.show()


plt.figure(figsize=(12,6))
plt.text(1,1,'1',ha='right',va= 'bottom',fontsize=19)
plt.text(5,5,'3',ha='right',va= 'bottom',fontsize=19)
plt.text(9,5,'2',ha='right',va= 'bottom',fontsize=19)
plt.text(2,2,'4',ha='right',va= 'bottom',fontsize=19)
plt.text(3,7,'5',ha='right',va= 'bottom',fontsize=19)
plt.scatter([1,2,3,5,9],[1,2,7,5,5],color='red',s=50)

p1 = plt.plot(list1,list2,color='#836FFF',linewidth=2,label='raw data')

p2 = plt.plot([1,2,3,5,9],[1,2,7,5,5],'--',color='#76EE00',linewidth=2,label='segment line')
plt.grid(axis="y")
font2 = {'family': 'Times New Roman',
         'weight': 'normal',
         'size': 30,
         }

plt.legend([p1,p2],labels=['raw_data_line','segment_line'],prop={'size':20},loc='lower right')
plt.xticks(list(range(10)),['','1', '2', '3', '4', '5', '6', '7', '8', '9'],fontsize=20)
plt.yticks(list(range(8)),['','1', '2', '3', '4', '5', '6', '7'],fontsize=20)

plt.show()
```
